Ein interessanter [Endpoint](https://data.nasa.gov/developer/external/planetary/#apod) ist der für das [Astronomy Picture of the Day](http://apod.nasa.gov/apod/astropix.html). Hier mal ein Billo-Beispiel:

Das [requests](http://docs.python-requests.org/en/latest/)-Modul kann man nicht genug loben. Es kann auch gleich das JSON parsen.

```python
import requests
```

Hier bekommen wir die Daten für das Bild an meinem Hochzeitstag:

```python
r = requests.get('https://api.data.gov/nasa/planetary/apod?date=2011-09-10&api_key=DEMO_KEY')
```

```python
r.json()
```

Neben der Beschreibung bekommen wir auch die URL für das Bild. Mal anzeigen lassen:

```python
from IPython.display import Image
Image(r.json()['url'])
```
